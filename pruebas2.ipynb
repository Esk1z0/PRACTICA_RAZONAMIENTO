```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:1234/v1",
    api_key="lm-studio"
)

def decide(state_text):
    r = client.chat.completions.create(
        model="meta-llama-3.1-8b-instruct",
        messages=[
            {"role": "system", "content": "Responde SOLO en JSON {action,args,confidence}"},
            {"role": "user", "content": state_text}
        ],
        temperature=0.2
    )
    return r.choices[0].message.content
```

```python
decide("El robot tiene una pared delante , decide accion")
```
